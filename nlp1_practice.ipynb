```python
import numpy as np
import pandas as pd
import re
import nltk 
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv(r"C:\Users\SSS\Downloads\bbc-text.csv")
df
```

```python
stp=stopwords.words("english")
```

```python
stp.remove("not")
```

```python
"not" in stp
```

```python
lt=[]
for word in df["text"]:
    l=[]
    ls=word.lower()
    ln=re.sub("[^a-zA-Z]"," ",ls)
    for s in word_tokenize(ln):
        if s in stp or len(s)<=2:
            pass
        else:
            ps=PorterStemmer().stem(s)
            l.append(ps) 
    lt.append(" ".join(l))    
```

```python
data=pd.DataFrame({"cl_data":lt})
data
```

```python
cv=CountVectorizer()
```

```python
dt=cv.fit_transform(data["cl_data"])
```

```python
cv.vocabulary_
```

```python
pd.DataFrame(dt.toarray(),columns=sorted(cv.vocabulary_.keys()))
```
